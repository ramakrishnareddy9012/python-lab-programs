1.1) Write a program that asks the user for a weight in kilograms and converts it to
pounds. There are 2.2 pounds in a kilogram.

```python
kilograms = float(input("Enter weight in kilograms: "))
pounds = kilograms * 2.2
print(f"{kilograms} kilograms is equal to {pounds} pounds")
```

1.5) Write a Python program to calculate the nth Fibonacci number using a function.

```python
def fibonacci(n):
  """Calculates the nth Fibonacci number."""
  if n <= 1:
    return n
  else:
    return(fibonacci(n-1) + fibonacci(n-2))

num = int(input("Enter a non-negative integer to find its Fibonacci number: "))

if num < 0:
  print("Please enter a non-negative integer.")
else:
  print(f"The {num}th Fibonacci number is: {fibonacci(num)}")
```
