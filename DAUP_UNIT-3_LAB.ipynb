b. To test whether none of the elements of a given array is zero.

```python
import numpy as np

arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([1, 2, 3, 0])

print("\nAll elements are non-zero:", np.all(arr1))
print("\nAll elements are non-zero:", np.all(arr2))
```

c. To create an element-wise comparison (greater, greater_equal, less and less_equal,
equal, equal within a tolerance) of two given arrays.

```python
import numpy as np

a = np.array([18, 17, 19])
b = np.array([7, 1, 45])

print("Greater:", np.greater(a, b))                 #[ True  True False]
print("Greater_Equal:", np.greater_equal(a, b))     #[ True  True False]
print("Less:", np.less(a, b))                       #[False False  True]
print("Less_Equal:", np.less_equal(a, b))           #[False False  True]
print("Equal:", np.equal(a, b))                     #[False False False]

#Check if arrays are almost equal (within tolerance).
c = np.array([1.00001, 2.00001])
d = np.array([1.00002, 2.00002])

print("Allclose:", np.allclose(c, d))               #True (within default tolerane)
```

3.2) Write a NumPy program using NumPy methods max, min, argmax, argmin,
argmax, repr, count, bincount, unique.

a. To extract all numbers from a given array which are less and greater than a specified
number.

```python
x = np.array([30, 40, 50, 18, 17, 19])

#Extract numbers greater than 20
print("Numbers greater than 20:", x[x > 20])

#Extract numbers less than 30
print("Numbers less than 30:", x[x < 30])
```

b. To find the indices of the maximum and minimum numbers along the given axis of an
array.

```python
import numpy as narayana

n_arr = narayana.array([[18, 17, 19], [7, 1, 45]])

#Index of max along axis 0 (column_wise)
print("Argmax (axis=0):", narayana.argmax(n_arr, axis=0))      #[0 0 1]

#Index of min along axis 1 (row_wise)
print("Argmin (axis=1):", narayana.argmin(n_arr, axis=1))      #[1 1]
```

```python
#Repr, Bincount and Unique

import numpy as n
x = n.array([2, 5, 6, 3, 6, 1, 5, 9])

print("Repr:", repr(x))
print("Bincount:", n.bincount(x))
print("Unique:", n.unique(x))
```
