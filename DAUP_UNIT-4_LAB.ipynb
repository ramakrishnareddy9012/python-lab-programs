4.1) Pandas DataSeries:

i)Write a Pandas program to create and display a one-dimensional array-like object
containing an array of data using Pandas module.

```python
import pandas as pd
import numpy as np

s = pd.Series([1, 3, 5, np.nan, 6, 8])
print("Pandas Series:")
print(s)
```

ii)Write a Pandas program to convert a Panda module Series to Python list and it's typ–µ.

```python
import pandas as pd
import numpy as np

list_from_series = s.tolist()

print("Python list from Series:")
print(list_from_series)

print("Type of the list:")
print(type(list_from_series))
```

i)Write a Pandas program to create and display a DataFrame froma specified dictionary
data which has the index labels.

```python
import pandas as pd
import numpy as np

exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael',
                      'Matthew', 'Laura', 'Kevin', 'Jonas'],
             'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
             'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
             'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

df = pd.DataFrame(exam_data, index=labels)

print(df)
```

ii)Write a Pandas program to change the name 'James' to 'Suresh' in name column of the
DataFrame.

```python
df['name'] = df['name'].replace('James', 'Suresh')

print("\nDataFrame after changing 'James' to 'Suresh':")
print(df)
```
